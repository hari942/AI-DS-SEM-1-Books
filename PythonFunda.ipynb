```python
import time
def timer(func):
  def wrapper(*args,**kwargs):
    start=time.time()
    result=func(*args,**kwargs)
    print("Estimated time",time.time()-start)
    return result
  return wrapper

@timer
def compute():
  time.sleep(1)

compute()
```

```python
def fibonacci(n):
  a,b=0,1
  for _ in range(n):
    yield a
    a,b=b,a+b

for num in fibonacci(5):
  print(num)
```

```python
import asyncio

async def fetch_all(name):
  await asyncio.sleep(1)
  return f"Data{name}"

async def main():
  result=await asyncio.gather(
      fetch_all("1"),
      fetch_all("2"),
      fetch_all("3")
  )
  print(result)
await (main())
```
